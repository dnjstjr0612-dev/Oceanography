## How do we describe and measure salinity
- `Absolute Salinty($S_A$)` represents, to the best available accuracy, the mass fraction of dissolved solute in so-called Standard Seawater with the same density as that of the sample. 
- $S_A$ is not measured directly. In practice, salinity is inferred from the conductivity of seawater (which can be measured _in-situ_ with electronic instruments) and reported as __"practical salinity"__ $S_P$

## Hydrostatic Balance ##

How much pressure does the water column exert? In other words, what is the weight of 1 square meter of ocean?

$$
p(z) = g \int_z^0 \rho(z') dz'
$$

The pressure at depth $z$ is given by the mass of water above. We can re-express this as

$$
\boxed{\frac{\partial p}{\partial z} = - g \rho}
$$

This equation is known as the __hydrostatic balance__.

```python
z = np.arange(0, -5000, -1)
z_lat_0 = xr.DataArray(z, dims=['p'], coords={'p':gsw.p_from_z(z, lat=0)}, name='z')
z_lat_70 = xr.DataArray(z, dims=['p'], coords={'p':gsw.p_from_z(z, lat=70)}, name='z')
```

## Temperature and Heat Content ##

In general, the thermal energy of a material is given by

$$
U = \int \int \int \rho c_p T dv
$$

where $\rho$ is the density, $c_p$ is the heat capacity, and $T$ is the temperature. The units of energy are Joules

Let's compare the heat that can be stored by the ocean and the atmosphere. The following table gives values for air and seawater near sea level. 

|Parameter|$\rho$ (kg m$^{-3}$)|$c_p$ (J K$^{-1}$ kg$^{-1}$)|
|--|--------|-----------------------|
|__Atmosphere__ | 1.3 | 1004  |
|__Ocean__ | 1023 | 4218 |

_Question:_ How much more heat can be stored in 1 square meter of ocean than one square meter of air?
$$
\frac{Q_{\text{ocean}}}{Q_{\text{air}}}
=
\frac{1023 \times 4218}{1.3 \times 1004}
\approx 3.3 \times 10^3
$$

```python
t0 = 10 # initial T [C]
p0 = 0  # initial P [dbar]
sa0 = 35  # initial S_a
ct0 = gsw.CT_from_t(sa0, t0, p0) # calculate initial `C`onservaive `T`emperature 열온도?
p_new = 4000 # new pressure
t_new = gsw.t_from_CT(sa0, ct0, p_new) # new T

# In-situ temperature
print(f'Initial temperature: {t0:.3f}, conservative temperature: {ct0:.3f}')

# Conservative temperature
print(f'Final   temperature: {t_new:.3f}, conservative temperature: {ct0:.3f}')
```

1. 초기 해수(s, t, p)에서 CT 고정
2. 압력 염분 바꾸면서 in-situ 온도가 얼마나 변하는지 2D map으로 그리기
3. 초기 온도별로 여러 장 생성

```python
# 0. initial 값 입력 -> filled contour plot 객체 출력
def change_in_temperature(init_sa, init_t, init_p):
    # 1. CT 계산 후 ct0 고정. heat content 상태 관찰
    ct0 = gsw.CT_from_t(init_sa, init_t, init_p)  # conversed
    
    # 2. 격자 만들기 (Sa축, P축, 100*100)
    sa_min, sa_max, p_min, p_max = 30, 40, 0, 6000
    sa_range = np.linspace(sa_min, sa_max, 100)
    p_range = np.linspace(p_min, p_max, 100)
    
    # 3. CT 고정한채로 in-situ T 복원.
    # shape (1, 100) * (100, 1) -> t를 (100, 100)
    t = gsw.t_from_CT(sa_range[np.newaxis], ct0, p_range[:, np.newaxis])
    
    # 4. delta_t 계산. 
    # 모든 격자점에서 '초기 온도에 비해, 같은 ct를 유지하면서 염분밀도 바꾸면 온도 얼마나 달라지나?'
    delta_t = t - init_t

    # 5. 데이터에 좌표 붙이기.
    da = xr.DataArray(delta_t, dims = ['P', 'S_a'],
                      coords = {'S_a':sa_range, 'P':p_range})
    
    # 6. holoviews 이미지 만든 후, 색 범위 조정
    img = hv.Image(da, label="Temperature Change")
    # 지배적인 항은 압력 증가(-> adiabatic heating)이며, 경험적으로 1도/4000dbar니까.. 1.5도 충분
    delta_t_range = (-1.5, 1.5)
    img = img.redim.range(z=delta_t_range)
    
    # 7. 등치선 + filled contour
    levels = np.arange(*(delta_t_range + (0.1,) )) 
    contours = hv.operation.contours(img, levels=levels, overlaid=False, filled=True)
    return contours
```

## Enthalpy and Conservative Temperature ##

__Enthalpy 엔탈피__ is represented by $H$ and definition is $$ H = U + pV $$ which U : internal energy, and pV : pressure-volume work. `Total energy`, [J]

__Specific enthalpy 비엔탈피__ is represented by $h$ and has units of J/kg. A change in $h$ is equal to the change in the internal energy of the system (thermal and chemical), plus the pressure work that the system does on its surroundings (compressibility), at constant pressure.
$$ h = \frac{H}{m} $$

__Potential enthalpy__ $h_0$ is the enthalpy a water parcel _would have_ if brought adiabatically to surface pressure.

The __conservative temperature__ is defined in TEOS-10 as

$$ \Theta = \frac{h_0}{c_p^0} $$

where $c_p^0$ is a _constant scale factor_. The value of $c_p^0$ is 3991.868 J kg$^{-1}$ K$^{-1}$.

| 개념                                | 의미              | 단위          |
| --------------------------------- | --------------- | ----------- |
| Enthalpy (H)                      | 전체 에너지          | J           |
| Specific enthalpy (h)             | kg당 에너지         | J kg(^{-1}) |
| Potential enthalpy (h_0)          | 압력 효과 제거한 에너지   | J kg(^{-1}) |
| Conservative Temperature (\Theta) | 열 함량을 나타내는 ‘온도’ | K (or °C)   |


A more complete measure of heat content of the ocean is

$$  \iiint \rho h_0 dv = c_p^0 \iiint \rho \Theta dv $$

This change in this quantity is proportional to the heat flux through the ocean surface. Furthermore $\Theta$ is _conserved_ following the fluid flow. (More on this to come.)

Prior to TEOS-10, __potential temperature__($\theta$) was the preferred temperature variable!

$\theta$ is the temperature that the water would have if it were brought adiabatically to the surface. Potential temperature was abandoned because it is not a conserved quantity and is not proportional to the heat content of the fluid.

potential temperature는 단열 이동에선 보존되지만, mixing이 일어나면 정확히 보존되지 않는다는 단점이 있고, 특히 같은 $\theta$를 가진 두 해수가 염분이 다르다면 실제 enthalpy가 다를 수 있다. 즉, 온도 변수인데 열을 제대로 대표하지 못한다는 취약점이 존재한다.

따라서 potential enthalpy를 통해 conservative temperature를 활용한다.

## Potential Temperature vs Conservative Temperature

Given an initial seawater state:

* Absolute Salinity: $S_A$
* In-situ temperature: $t$
* Pressure: $p$

---

### 1. Conservative Temperature

In TEOS-10, the **Conservative Temperature (CT)** is defined as:

$$
CT = \frac{h_0(S_A, t, p)}{c_p^0}
$$

where:

* $h_0$ is the **potential enthalpy** (enthalpy after adiabatic transport to the surface)
* $c_p^0 = 3991.868\ \mathrm{J\ kg^{-1}\ K^{-1}}$ is a constant reference heat capacity

Because $c_p^0$ is constant,

$$
CT \propto h_0
$$

and therefore CT is **linearly proportional to heat content**.

---

### 2. Potential Temperature

The **Potential Temperature** $\theta$ is defined implicitly by:

$$
h_0(S_A, t, p) = h(S_A, \theta, p_{ref})
$$

with the reference pressure:

$$
p_{ref} = 0\ \mathrm{dbar}
$$

In differential form:

$$
d\theta = \frac{dh}{c_p(S_A, T, p)}
$$

where the specific heat capacity depends on state:

$$
c_p = c_p(S_A, T, p)
$$

---

### 3. Key Difference: Linearity vs Nonlinearity

For Conservative Temperature:

$$
CT = \frac{h_0}{c_p^0}
$$

This relationship is **linear in enthalpy**.

For Potential Temperature:

$$
\theta = \int \frac{1}{c_p(S_A, T, p)} , dh
$$

This relationship is **nonlinear in enthalpy**.

---

### 4. Physical Interpretation

Since heat capacity increases with pressure:

$$
c_p(S_A, T, p) > c_p^0
$$

for deep-ocean conditions, which implies:

$$
\theta > CT
$$

Thus, potential temperature systematically **underestimates cooling** compared to Conservative Temperature.

---

### 5. Mixing Behavior

For two parcels with masses $m_1$ and $m_2$:

Enthalpy conservation:

$$
h_{mix} = \frac{m_1 h_1 + m_2 h_2}{m_1 + m_2}
$$

Conservative Temperature:

$$
CT_{mix} = \frac{m_1 CT_1 + m_2 CT_2}{m_1 + m_2}
$$

Potential Temperature:

$$
\theta_{mix} \neq \frac{m_1 \theta_1 + m_2 \theta_2}{m_1 + m_2}
$$

---

### 6. Conclusion

* Potential temperature $\theta$ is not proportional to heat content
* Conservative Temperature $CT$ is directly proportional to potential enthalpy
* TEOS-10 adopts $CT$ as the energetically consistent temperature variable
