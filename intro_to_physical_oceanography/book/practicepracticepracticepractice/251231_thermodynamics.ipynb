## How do we describe and measure salinity
- `Absolute Salinty($S_A$)` represents, to the best available accuracy, the mass fraction of dissolved solute in so-called Standard Seawater with the same density as that of the sample. 
- $S_A$ is not measured directly. In practice, salinity is inferred from the conductivity of seawater (which can be measured _in-situ_ with electronic instruments) and reported as __"practical salinity"__ $S_P$

## Pressure
- An isotropic force per square meter.
Units [Pa]
$$
Pa = \frac{N}{m^2} = \frac{kg}{m s^2}
$$

- Sea pressure $p$ is absolute pressure minus a constant reference atmos pressure:
$$ p = P - P_0 $$



## Hydrostatic Balance ##

How much pressure does the water column exert? In other words, what is the weight of 1 square meter of ocean?

$$
p(z) = g \int_z^0 \rho(z') dz'
$$

The pressure at depth $z$ is given by the mass of water above. We can re-express this as

$$
\boxed{\frac{\partial p}{\partial z} = - g \rho}
$$

This equation is known as the __hydrostatic balance__.

```python
z = np.arange(0, -5000, -1)
z_eq = xr.DataArray(z, dims=['p'], coords = {'p' : gsw.p_from_z(z, lat=0)}, name='z')
z_hi = xr.DataArray(z, dims=['p'], coords = {'p' : gsw.p_from_z(z, lat=70)}, name='z')
z_eq.sel(p=1008, method = 'nearest')
```

```python
# 1. ë°ì´í„° ì„¤ì •
lat = np.linspace(-90, 90, 181)  # ìœ„ë„: ë‚¨ê·¹(-90) ~ ë¶ê·¹(90)
depths = [0, 1000, 2000, 3000, 4000] # ë³´ê³  ì‹¶ì€ ìˆ˜ì‹¬ ë¦¬ìŠ¤íŠ¸

# 2. ê·¸ëž˜í”„ ê·¸ë¦¬ê¸° ì¤€ë¹„
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# ìƒ‰ìƒ ë§µ ì„¤ì • (ê¹Šì´ì— ë”°ë¼ ìƒ‰ ë‹¤ë¥´ê²Œ)
colors = plt.cm.viridis(np.linspace(0, 1, len(depths)))

for z_val, color in zip(depths, colors):
    z = -z_val  # gsw í•¨ìˆ˜ì—ëŠ” ê¹Šì´ê°€ ìŒìˆ˜(-)ë¡œ ë“¤ì–´ê°€ì•¼ í•¨
    
    # í•´ë‹¹ ê¹Šì´ì—ì„œ ìœ„ë„ë³„ ì••ë ¥ ê³„ì‚°
    p = gsw.p_from_z(z, lat)
    
    # (1) ì ˆëŒ€ ì••ë ¥ ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°
    ax1.plot(lat, p, label=f'{z_val}m', color=color, linewidth=2)
    
    # (2) ì••ë ¥ ì°¨ì´ ê·¸ëž˜í”„ ê·¸ë¦¬ê¸° (ê° ìœ„ë„ì˜ ì••ë ¥ - ì ë„(0ë„)ì—ì„œì˜ ì••ë ¥)
    p_eq = gsw.p_from_z(z, 0) # ì ë„ ì••ë ¥
    p_diff = p - p_eq         # ì°¨ì´ ê³„ì‚°
    
    ax2.plot(lat, p_diff, label=f'{z_val}m (Max diff: {np.max(p_diff):.1f} dbar)', color=color, linewidth=2)

# 3. ê·¸ëž˜í”„ ê¾¸ë¯¸ê¸°

# ì²« ë²ˆì§¸ ê·¸ëž˜í”„: ì ˆëŒ€ ì••ë ¥
ax1.set_ylabel('Pressure [dbar]', fontsize=12)
ax1.set_title('Pressure vs Latitude at Constant Depths', fontsize=14, fontweight='bold')
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend(loc='upper center')

# ë‘ ë²ˆì§¸ ê·¸ëž˜í”„: ì ë„ ëŒ€ë¹„ ì••ë ¥ ì°¨ì´ (ë³€í™”ëŸ‰ì„ í™•ëŒ€í•´ì„œ ë³´ê¸° ìœ„í•¨)
ax2.set_ylabel('Pressure Anomaly (P - P_equator) [dbar]', fontsize=12)
ax2.set_xlabel('Latitude [degrees]', fontsize=12)
ax2.set_title('Pressure Difference from Equator', fontsize=14, fontweight='bold')
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend(loc='upper center')

# xì¶• ì„¤ì • (ìœ„ë„ í‘œì‹œ)
plt.xlim(-90, 90)
plt.xticks(np.arange(-90, 91, 30))

plt.tight_layout()
plt.show() # í˜¹ì€ plt.savefig('pressure_vs_lat.png')
```
